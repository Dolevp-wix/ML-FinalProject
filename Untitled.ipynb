```python
def gatherHistoricalDataForKey(apiKey, startDate, timeDelta = timedelta(days=20, hours=0)):
    url = "https://visual-crossing-weather.p.rapidapi.com/history"
    headers = {
        "X-RapidAPI-Key": apiKey,
        "X-RapidAPI-Host": "visual-crossing-weather.p.rapidapi.com"
    }
    start = startDate
    end = startDate - timeDelta
    dataFrames = []
    while True:
        querystring = {
            "startDateTime":f"{start.isoformat()}T00:00:00",
            "aggregateHours":"1",
            "location":"Jerusalem,ISR",
            "endDateTime":f"{end.isoformat()}T00:00:00",
            "unitGroup":"us",
            "dayStartTime":"00:00:00",
            "contentType":"csv",
            "dayEndTime":"23:00:00",
            "shortColumnNames":"0"
        }
        response = requests.request("GET", url, headers=headers, params=querystring)
        if response.status_code != 200:
            print(f'finished batch for key {apiKey} and found {len(dataFrames)} rows and finished on end date {end}')
            break
        df = pd.read_csv(io.StringIO(response.content.decode('utf-8')))
        dataFrames.append(df)
        start = end
        end = end - timeDelta
    return dataFrames, end
```

```python
def gatherHistory(apiKeys, startDate):
    dataFrames = []
    start = startDate
    for apiKey in apiKeys:
        currentBatchDataFrames, end = gatherHistoricalDataForKey(apiKey, start)
        start = end
        dataFrames += currentBatchDataFrames
    return dataFrames, end
```
