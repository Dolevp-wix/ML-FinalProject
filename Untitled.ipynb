```python
import requests
import json
import numpy as np
import pandas as pd
from pandas import json_normalize
from datetime import date
from datetime import timedelta
import io
import time
```

```python
def gatherHistoricalDataForKey(apiKey, endDate, timeDelta = timedelta(days=19, hours=0)):
    url = "https://visual-crossing-weather.p.rapidapi.com/history"
    headers = {
        "X-RapidAPI-Key": apiKey,
        "X-RapidAPI-Host": "visual-crossing-weather.p.rapidapi.com"
    }
    start = endDate - timeDelta
    end = endDate
    dataFrames = []
    while True:
        if (end <= date(1970, 1, 1)):
            break
        querystring = {
            "startDateTime":f"{start.isoformat()}T00:00:00",
            "aggregateHours":"1",
            "location":"Jerusalem,ISR",
            "endDateTime":f"{end.isoformat()}T00:00:00",
            "unitGroup":"us",
            "dayStartTime":"00:00:00",
            "contentType":"csv",
            "dayEndTime":"23:00:00",
            "shortColumnNames":"0"
        }
        response = requests.request("GET", url, headers=headers, params=querystring)
        if response.status_code != 200:
            print(f'finished batch for key {apiKey} and found {len(dataFrames)} rows and finished on end date {end} and responseCode is {response.status_code} content is {response.text}')
            break
        df = pd.read_csv(io.StringIO(response.content.decode('utf-8')))
        dataFrames.append(df)
        end = start
        start = end - timeDelta
        time.sleep(1)
    return dataFrames, end
```

```python
globalDataFrames = []
def gatherHistory(apiKeys, startDate):
    global globalDataFrames
    start = startDate
    for apiKey in apiKeys:
        currentBatchDataFrames, end = gatherHistoricalDataForKey(apiKey, start)
        start = end
        globalDataFrames += currentBatchDataFrames
    return globalDataFrames, end
```
